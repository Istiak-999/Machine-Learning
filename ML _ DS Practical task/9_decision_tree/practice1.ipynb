```python
import pandas as pd
df=pd.read_csv('salaries.csv')
df
```

```python
c_dummies=pd.get_dummies(df.company)
j_dummies=pd.get_dummies(df.job)
d_dummies=pd.get_dummies(df.degree)
```

```python
ex=pd.concat([c_dummies,j_dummies,d_dummies],axis='columns')
ex
```

```python
new_ex=ex.drop(["google","sales executive","masters"],axis='columns')
new_ex
```

```python
new_df=pd.concat([df,new_ex],axis='columns')
new_df
```

```python
final=new_df.drop(["company","job","degree"],axis='columns')
final
```

```python
input=final.drop(['salary_more_then_100k'],axis='columns')
input
```

```python
target=df.salary_more_then_100k
target
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(input,target)
```

```python
model.predict([[0,0,0,1,1]])
```
