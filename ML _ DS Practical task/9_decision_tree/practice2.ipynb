```python
import pandas as pd
df=pd.read_csv('salaries.csv')
df
```

```python
c_dummies=pd.get_dummies(df["company"])
j_dummies=pd.get_dummies(df["job"])
d_dummies=pd.get_dummies(df["degree"])
```

```python
temp=pd.concat([c_dummies,j_dummies,d_dummies],axis='columns')
new_temp=temp.drop(["google","sales executive","masters"],axis='columns')
new_temp
```

```python
ndf=pd.concat([df,new_temp],axis='columns')
ndf
```

```python
new_df=ndf.drop(["company","job","degree"],axis='columns')
new_df
```

```python
inputs=new_df.drop(['salary_more_then_100k'],axis='columns')
inputs
```

```python
result=new_df.salary_more_then_100k
result
```

```python
from sklearn.model_selection import train_test_split
inputs_train,inputs_test,result_train,result_test=train_test_split(inputs,result,test_size=0.3)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_train,result_train)
```

```python
result_test
```

```python
model.predict(inputs_test)
```

```python
model.score(inputs_test,result_test)
```

```python
import joblib
joblib.dump(model,'joblib_save')
```
