```python
import pandas as pd
df=pd.read_csv('titanic.csv')
df
```

```python
input=df.drop(["PassengerId","Name","SibSp","Parch","Ticket","Cabin","Embarked"],axis='columns')
input.head(496)
```

```python
dummies=pd.get_dummies(input.Sex)
dummies
```

```python
marge=pd.concat([input,dummies],axis='columns')
marge.head(10)
```

```python
final=marge.drop(["Sex","male"],axis='columns')
final.head(10)
```

```python
final['Age']=final['Age'].fillna(final['Age'].median())
inputs=final.drop(['Survived'],axis='columns')
survive=final.Survived
```

```python
from sklearn.model_selection import train_test_split
inputs_train,inputs_test,survive_train,survive_test=train_test_split(inputs,survive,test_size=0.2)
```

```python
inputs_train
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_train,survive_train)
```

```python
survive_test
#model.predict(survive_train)
```

```python
survive_train
```

```python
model.predict(inputs_test)
```

```python
model.score(inputs_test,survive_test)
```

```python
import joblib
joblib.dump(model,"joblib_save")
```
